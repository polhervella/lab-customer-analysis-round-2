```python
# 1.Show the dataframe shape

import pandas as pd

file1 = "marketing_customer_analysis.csv"
Mk_ca = pd.read_csv(file1)
Mk_ca
```

```python
#2.Standardize header names

Mk_ca.columns = Mk_ca.columns.str.upper().str.replace(" ","_")
Mk_ca
```

```python
# 3/4 Which columns are numerical and which ones categorical?

Mk_ca.dtypes
```

```python
#Check and deal with NaN values

Mk_ca.isna().sum()
```
